## Introduction

Welcome to this example notebook! This notebook demonstrates:

- Syntax highlighting

Let's get started with some Python code examples.


```python
# Simple Python example
def greet(name):
    return f"Hello, {name}!"

message = greet("World")
print(message)
```

## Data Structures

Here's an example working with Python data structures:


```python
# Working with lists and dictionaries
fruits = ['apple', 'banana', 'cherry']
fruit_colors = {
    'apple': 'red',
    'banana': 'yellow',
    'cherry': 'red'
}

for fruit in fruits:
    color = fruit_colors.get(fruit)
    print(f"{fruit.capitalize()} is {color}")
```

## Mathematical Operations

Let's perform some basic mathematical operations:


```python
import math

# Calculate the area of a circle
radius = 5
area = math.pi * radius ** 2
circumference = 2 * math.pi * radius

print(f"Circle with radius {radius}:")
print(f"Area: {area:.2f}")
print(f"Circumference: {circumference:.2f}")
```
